# The Aerosol Life Cycle

There are many types of aerosols in the atmosphere. Each type can be classified as Natural or Anthropogenic. Natural aerosols are created through various Earth processes. Examples include dust, sea salt, and smoke. Anthropogenic aerosols are formed through man-made events. Examples include sulfates and black carbon.

The distribution of an aerosol depends on many factors. Below are mass concentration charts of three different types of aerosols, particulate matter 2.5, dust, and black carbon:

![Aerosol Distribution Comparison](aerosol_comparison.png)

Particulate matter 2.5 represents all particles under 2.5 microns in diameter. In remote, desert regions, the majority of aerosols are natural aerosols, such as dust. In heavily populated regions, most aerosols are classified as anthropogenic.

Aerosols play a large role in how radiation interacts with the Earth. Aerosol optical depth measures how much of a solar radiation beam is extinguished by aerosols. Below are two different model estimates of aerosol optical depth around the world. Lighter colors indicate more solar radiation is able to reach Earth's surface, while darker colors indicate less:

![Model Comparison](aod_model_comparison.png)

Notice how regions with more aerosols allow less solar radiation to the Earth's surface, whether those aerosols are natural or anthropogenic. Also, observe the many differences between the two models. Because models are estimates based on equations and initial conditions, they can produce a variety of different outputs.
